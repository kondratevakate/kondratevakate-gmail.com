Function for `CanICA()` feeding and transformation.

## CONTROLS

```python
from nilearn.decomposition import CanICA

model = CanICA(n_components=70,
                 smoothing_fwhm= 6.,
                 random_state= 0, 
                 n_jobs= -1
                )
fit_list = list_rest['CONTROL']

print ('Fitting the model ...')
t0 = time()
model.fit(fit_list)
print ('time() on fitting:  ', time() - t0)
```

```python
from nilearn.image import smooth_img

def can_ica_projection(list_img, model, target =''):
    resulting_array = []
    
    for i in list_img:
        print ('Dowloading image ', i)
        print ()
        t0 = time()
        img = smooth_img([i], 6)
        temp_transform = model.transform(img)
        
        t_shape = np.shape(temp_transform)[1]
        i_shape = np.shape(temp_transform)[2]
        
        resulting_array.append(np.array(temp_transform).reshape(t_shape *i_shape))
        print ('time() on fitting:  ', time() - t0)
        print ()
        
    result = pd.DataFrame(resulting_array)
    result['target'] = target
    
    return(result)
```

```python
bipolar_projection_rs = can_ica_projection(list_rest['BIPOLAR'], model = model, target ='BIPOLAR');
adhd_projection_rs = can_ica_projection(list_rest['ADHD'], model = model, target ='ADHD');
```

```python
schz_projection_rs = can_ica_projection(list_rest['SCHZ'], model = model, target ='SCHZ');
control_projection_rs= can_ica_projection(list_rest['CONTROL'], model = model, target = 'CONTROL')
```

```python
bipolar_projection_tb = can_ica_projection(list_bart['BIPOLAR'], model = model, target ='BIPOLAR');
adhd_projection_tb = can_ica_projection(list_bart['ADHD'], model = model, target ='ADHD');
```

```python
def tsne_plot (X , z, scatter, title = False):
    from sklearn.manifold import TSNE
    from sklearn.preprocessing import LabelEncoder
    import matplotlib.cm as cm
    
    le = LabelEncoder()
    z = le.fit_transform(z)
    tsne = TSNE(n_components=2, perplexity = 10, n_iter = 3000, learning_rate = 2, random_state = 42)
    T = tsne.fit_transform(X)

    plt.figure(figsize = (8, 8))

    for c_i, c in enumerate(set(z)):
        sub_sub_inds = np.where(z == c)[0]
        plt.scatter(T[sub_sub_inds, 0], 
                    T[sub_sub_inds, 1], 
                    s = scatter,
                    marker = 'o', alpha=0.45, edgecolor='k',
                    label = str(le.inverse_transform([c])[0]), 
                    color = cm.tab10(int(c_i)))

    lgd = plt.legend(loc = 2, bbox_to_anchor = (1, 1), markerscale = 2, fontsize = 14)
    plt.xlabel('Embedding axis 1', fontsize = 14)
    plt.ylabel('Embedding axis 2', fontsize = 14)
    if title:
        plt.savefig(title, dpi=400, bbox_extra_artists=(lgd,), bbox_inches='tight')
    return (T)
```

```python
T = tsne_plot(X = X, scatter = 20, z = data[data.columns[-1]] )
```

```python
def tsne_plot (X1, X2 , z1, z2,  title = False):
    from sklearn.manifold import TSNE
    from sklearn.preprocessing import LabelEncoder
    import matplotlib.cm as cm
    
    le = LabelEncoder()
    z = le.fit_transform(z1)
    tsne = TSNE(n_components=2, perplexity = 7, n_iter = 3000, learning_rate = 2, random_state = 42)
    T1 = tsne.fit_transform(X1)
    T2 = tsne.fit_transform(X2)

    plt.figure(figsize = (8, 8))

    for c_i, c in enumerate(set(z)):
        sub_sub_inds = np.where(z == c)[0]
        plt.scatter(T1[sub_sub_inds, 0], 
                    T1[sub_sub_inds, 1],
                    marker = 'o', alpha=0.45, edgecolor='k',
                    label = str(le.inverse_transform([c])[0]), 
                    color = cm.tab10(int(c_i)))
    
    z = le.fit_transform(z2)
    
    for c_i, c in enumerate(set(z)):
        sub_sub_inds = np.where(z == c)[0]
        plt.scatter(T2[sub_sub_inds, 0], 
                    T2[sub_sub_inds, 1], 
                    marker = 'd', alpha=0.45, edgecolor='k',
                    label = str(le.inverse_transform([c])[0]), 
                    color = cm.tab10(int(c_i)))

    lgd = plt.legend(loc = 2, bbox_to_anchor = (1, 1), markerscale = 2, fontsize = 14)
    plt.xlabel('Embedding axis 1', fontsize = 14)
    plt.ylabel('Embedding axis 2', fontsize = 14)
    if title:
        plt.savefig(title, dpi=400, bbox_extra_artists=(lgd,), bbox_inches='tight')
    return (T)
```

```python
X1 = data_rs[data_rs.columns[:-1]]
X1 = X1.T.dropna().T
z1 = data_rs[data_rs.columns[-1]]
z1 = z1 + '_rs'

X2 = data_tb[data_tb.columns[:-1]]
X2 = X2.T.dropna().T
z2 = data_rs[data_rs.columns[-1]]
z2 = z2 + '_tb'

T = tsne_plot(X1, X2 , z1, z2 )
```
